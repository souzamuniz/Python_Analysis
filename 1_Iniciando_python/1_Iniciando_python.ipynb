```python
# Primeiro instalamos e chamamos o pacote math, pois iremos utilizar a função sqrt
# import math  -- Caso queira chamar todo o pacote math
from math import sqrt  # Veja que por aqui, extraimos apenas a função desejada do pacote
```

```python
# Apresentção do código ao usuário e inserção de dados de entrada

print('Script de Resolução de Equação de Segundo Grau criado pelo grupo 9')

a = input('Por favor, insira o valor de a:  ');
b = input('Por favor, insira o valor de b:  ');
c = input('Por favor, insira o valor de c:  ');

a = float(a);
b = float(b);
c = float(c);
```

```python
# Caso o usuário insira um valor de a = 0, não teremos o termo quadráico e não se configurará como uma expressão de segundo grau
# Dessa forma se cobrará do usuário um termo válido
while a == 0:
        a = input('Por favor, insira um valor de "a" compatível a uma equação de segundo grau:  ');
        a = float(a);    
```

```python
# Cálculo do delta: 
aux = 1;

delta = (b**2) - (4*a*c)
if delta < 0:
    print("A Equação não tem raízes reais")
    aux = 0;
```

```python
# Calculando as raizes 
if aux:
    x1 = (-b + sqrt(delta))/(2*a)
    x2 = (-b - sqrt(delta))/(2*a)
    
    # Apresentação dos resultados
    print('As raízes reais da equação são:',round(x1,2),'e',round(x2,2))

```

### Podemos resumir o passo a passo em uma função

```python
# Desenvolvendo uma função chamada calc_eq_2_grau

def calc_eq_2_grau(a,b,c): # Definimos a função por meio do comando def
    
    from math import sqrt  # Importando a função sqrt do pacote math
    
    if a == 0:
            print('insira um valor de "a" compatível a uma equação de segundo grau')  # Caso nãos seja uma eq. de segundo grau, a função avisa ao usuário
    else:
            delta = (b**2) - (4*a*c)   # Cálculo do delta
            if delta < 0:
                print("A Equação não tem raízes reais")  # Caso não haja variáveis reais a função avisa ao usuário
            else:
                x1 = (-b + sqrt(delta))/(2*a)     # Cálculo das raízes
                x2 = (-b - sqrt(delta))/(2*a) 

                return(round(x1,2),round(x2,2))   # Arredondamos os valores finais para duas casas decimais por meio do comando round
           

        
```
