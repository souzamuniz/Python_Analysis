```python
# Primeiro instalamos e chamamos o pacote math, pois iremos utilizar a funo sqrt
# import math  -- Caso queira chamar todo o pacote math
from math import sqrt  # Veja que por aqui, extraimos apenas a funo desejada do pacote
```

```python
# Apresento do cdigo ao usurio e insero de dados de entrada

print('Script de Resoluo de Equao de Segundo Grau criado pelo grupo 9')

a = input('Por favor, insira o valor de a:  ');
b = input('Por favor, insira o valor de b:  ');
c = input('Por favor, insira o valor de c:  ');

a = float(a);
b = float(b);
c = float(c);
```

```python
# Caso o usurio insira um valor de a = 0, no teremos o termo quadrico e no se configurar como uma expresso de segundo grau
# Dessa forma se cobrar do usurio um termo vlido
while a == 0:
        a = input('Por favor, insira um valor de "a" compatvel a uma equao de segundo grau:  ');
        a = float(a);    
```

```python
# Clculo do delta: 
aux = 1;

delta = (b**2) - (4*a*c)
if delta < 0:
    print("A Equao no tem razes reais")
    aux = 0;
```

```python
# Calculando as raizes 
if aux:
    x1 = (-b + sqrt(delta))/(2*a)
    x2 = (-b - sqrt(delta))/(2*a)
    
    # Apresentao dos resultados
    print('As razes reais da equao so:',round(x1,2),'e',round(x2,2))

```

### Podemos resumir o passo a passo em uma funo

```python
# Desenvolvendo uma funo chamada calc_eq_2_grau

def calc_eq_2_grau(a,b,c): # Definimos a funo por meio do comando def
    
    from math import sqrt  # Importando a funo sqrt do pacote math
    
    if a == 0:
            print('insira um valor de "a" compatvel a uma equao de segundo grau')  # Caso nos seja uma eq. de segundo grau, a funo avisa ao usurio
    else:
            delta = (b**2) - (4*a*c)   # Clculo do delta
            if delta < 0:
                print("A Equao no tem razes reais")  # Caso no haja variveis reais a funo avisa ao usurio
            else:
                x1 = (-b + sqrt(delta))/(2*a)     # Clculo das razes
                x2 = (-b - sqrt(delta))/(2*a) 

                return(round(x1,2),round(x2,2))   # Arredondamos os valores finais para duas casas decimais por meio do comando round
           

        
```
